# Setup environment

 this project use uv manage the virtual environment, but uv is not working very well with pytorch, we need additional steps setup the cuda & pytorch environment, so we can use cuda-accelerated pytorch in this project.

the environment we are using:

- **PyTorch**: 2.9.0+cu130
- **TorchVision**: 0.24.0+cu130
- **CUDA Version**: 13.0

First, we add the following dependencies to the `pyproject.toml` file:

```toml

[tool.uv.sources]
torch = [
    { index = "pytorch-cu130", marker = "sys_platform == 'windows'"},
    { index = "pytorch-cpu", marker = "sys_platform != 'windows'"},
]

torchvision = [
    { index = "pytorch-cu130", marker = "sys_platform == 'windows'"},
    { index = "pytorch-cpu", marker = "sys_platform != 'windows'"},
]

[[tool.uv.index]]
name = "pytorch-cpu"
url = "https://download.pytorch.org/whl/cpu"
explicit = true

[[tool.uv.index]]
name = "pytorch-cu130"
url = "https://download.pytorch.org/whl/cu130"
explicit = true

[project.optional-dependencies]
cuda = [
    "torch>=2.9.0; sys_platform == 'windows'",
    "torch>=2.9.0; sys_platform != 'windows'",
    "torchvision>=0.24.0; sys_platform == 'windows'",
    "torchvision>=0.24.0; sys_platform != 'windows'",
]
```

Then we run the following command to install the cuda-supported pytorch:

```bash
uv pip install torch==2.9.0+cu130 torchvision==0.24.0+cu130 --index-url https://download.pytorch.org/whl/cu130
```
